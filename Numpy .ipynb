##  Plot二维图

```python
%matplotlib inline
a = linspace(0, 2*pi, 50)
b = sin(a)
plot(a,b) 
# 清除已有的图像的使用
# clf()
figure()
xlabel('radians')
ylabel('amplitude', fontsize='large')
title('Sin(x)')
plot(b) 
# 关闭当前图像
# close()
# 关闭所有
# close('all')
```

```python
plot(a, b, label="radians")
# 标签是对应的plot所画的线的
legend()
```

```python
mark = b > 0
print(mark)
print(a)
print(a[mark], b[mark])
# print(a & b> 0)
plot(a[mark], b[mark], 'bo')
```

```python
%matplotlib inline 
a = linspace(0, 2*pi, 50)
x = sin(a)
y = cos(a)
plot(x, label='111')
# hold()
plot(y, label='222')
legend()
# legend('sin(x) cos(x)')
# 会重新覆盖掉
legend(['sin(x) cos(x)'])
# legend(['sin', 'cos'])
title("sin&&cos")
# 显示网格
grid()
```

```python
%matplotlib inline 
a = linspace(0,10,50)
y = a**2
plot(a,'r-^')
figure()
plot(a,y, 'b-o')
```

```python
%matplotlib inline 
a = linspace(0, 10, 50)
subplot(1,3,1)
subplot?
plot(a)
xlabel("x")
ylabel("y=x/40")
title("x,y")
subplot(1,3,2)
plot(a,a*a)
grid()
subplot(1,3,3)
plot(a,a**3)
plot(a,a**3)
```

## Scatter 散点图

```python
x = rand(200)
y = rand(200)
color = rand(200) * 30
size = rand(200)
scatter(x, y, size, color)
# 显示颜色条
colorbar()
```

```python
t = linspace(0, 2*pi, 40)
# print(t)
# print(sin(t))
x = sin(t)
y = cos(t)
figure()
plot(x)
figure()
plot(y)
```
